#### Tipos de gráficos para analizar la distribución de los datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importando el data set de iris que viene precargado en seaborn
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Scatterplot Diagrama de dispersión
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.show()
```

```python
# Joint plot o unión de dos gráficos
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.show()
```

```python
# Box plot o diagrama de cajas y bigites
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.show()
```

```python
sns.jointplot(data=iris, x='petal_length', y='sepal_length', hue= ('species'=='setosa'))
plt.show()
```

```python
# Bar plot o Gráfico de barras
sns.barplot(x='species', y='sepal_length', data = iris)
plt.show()
```

```python
sns.jointplot(data=iris, x='petal_length', y='sepal_length')
plt.show()
```
